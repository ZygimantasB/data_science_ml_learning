```python
def yelp(price):
  if price < 10:
    return '$'
  elif 10 <= price < 30:
    return '$$'
  else:
    return '$$$'
```

```python
df['Expensis'] = df['total_bill'].apply(yelp)
```

```python
def quality(total_bill, tip):
  if tip/total_bill > 0.25:
    return 'Generous'
  else:
    return 'Other'
```

```python
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
df
```

```python
df['tip_quality1'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
df.drop(labels='tip_quality1', axis=1, inplace=True)
```
