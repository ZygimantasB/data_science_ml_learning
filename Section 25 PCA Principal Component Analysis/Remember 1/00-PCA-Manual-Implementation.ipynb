```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../DATA/cancer_tumor_data_features.csv')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
scaled_X -= scaled_X.mean(axis=0)
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
pd.DataFrame(covariance_matrix, columns=df.columns, index=df.columns)
```

```python
eigen_values, eigen_vectors = np.linalg.eigh(covariance_matrix)
```

```python
print(pd.DataFrame(eigen_values, index=df.columns, columns=['eigen_values']))
```

```python
pd.DataFrame(eigen_vectors, index=df.columns, columns=df.columns).to_dict()
```

```python
pd.DataFrame(eigen_vectors, index=df.columns, columns=df.columns)
```

```python
num_components = 4
```

```python
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
pd.DataFrame(sorted_key, index=range(1, num_components+1), columns=['sorted_key'])
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_key], eigen_vectors[:, sorted_key]
```

```python
pd.DataFrame(eigen_values, index=range(1, num_components+1), columns=['eigen_values'])
```

```python
pd.DataFrame(eigen_vectors, index=df.columns, columns=range(1, num_components+1))
```

```python
principal_components = np.dot(scaled_X, eigen_vectors)
```

```python
principal_components
```

```python
principal_component = np.dot(scaled_X, eigen_vectors)
```

```python
principal_component
```

```python
%matplotlib notebook
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0],principal_components[:,1], c='r', edgecolors='k', alpha=0.6, s=50)
plt.grid()
plt.title('First two principal components')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```
