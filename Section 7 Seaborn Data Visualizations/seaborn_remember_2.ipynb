```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_path = r'F:\DataSpell\data_science_ml_learning\Section 7 Seaborn Data Visualizations\data\dm_office_sales.csv'
```

```python
df = pd.read_csv(df_path)
```

```python
df
```

```python
sns.scatterplot(x='sales', y='salary', data=df)
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df, x='salary', y='sales', hue='division')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df, x='salary', y='sales', hue='work experience')
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df, x='salary', y='sales', hue='work experience', palette='coolwarm')
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df, x='salary', y='sales', size='work experience')
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='salary', y='sales', data=df, s=200)
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df, x='salary', y='sales', s=200, alpha=1.0)
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df, x='salary', y='sales', style='level of education')
```
