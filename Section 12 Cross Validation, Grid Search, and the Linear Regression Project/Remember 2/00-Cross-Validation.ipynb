```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../DATA/Advertising.csv")
```

```python
df
```

```python
X = df.drop('sales', axis=1)
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
```

```python
print(f'MAE: {MAE}, MSE: {MSE}, RMSE: {RMSE}')
```

```python
model = Ridge(alpha=1)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
```

```python
print(f'MAE: {MAE}, MSE: {MSE}, RMSE: {RMSE}')
```

# Cross Validation with cross validate

```python
X = df.drop('sales', axis=1)
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
scores = cross_validate(
    model, X_train, y_train,
    n_jobs=-1,
    scoring=['neg_mean_absolute_error', 'neg_mean_squared_error', 'max_error'],
    cv=5,
    return_train_score=True,
)
```

```python
pd.DataFrame(scores)
```

```python
pd.DataFrame(scores).mean()
```

```python
model = Ridge(alpha=1)
```

```python
scores = cross_validate(
    model, X_train, y_train,
    n_jobs=-1,
    scoring=['neg_mean_absolute_error','neg_mean_squared_error','max_error'],
    cv=5,
)
```

```python
pd.DataFrame(scores)
```

```python
pd.DataFrame(scores).mean()
```

```python
model.fit(X_train, y_train)
```

```python
y_final_tst_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, y_final_tst_pred)
MSE = mean_squared_error(y_test, y_final_tst_pred)
RMSE = np.sqrt(MSE)
```

```python
print(f'MAE: {MAE}, MSE: {MSE}, RMSE: {RMSE}')
```
