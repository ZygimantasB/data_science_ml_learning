```python
df_path = r'F:\DataSpell\data_science_ml_learning\Section 5 Pandas\data_frame\Sales_Funnel_CRM.csv'
```

```python
df = pd.read_csv(df_path)
```

```python
licenses = df[['Company','Product','Licenses']]
licenses
```

```python
pd.pivot(licenses, index='Company', columns='Product', values='Licenses')
```

```python
pd.pivot_table(df, index='Company', aggfunc='sum')
```

```python
pd.pivot_table(df, index='Company', aggfunc='sum', values=['Licenses', 'Sale Price'])
```

```python
pd.pivot(df, index='Company', columns='Product', values='Licenses')
```

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],aggfunc='sum')
```

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],
               aggfunc=['sum', 'mean'],fill_value=0)
```

```python
pd.pivot_table(df,index=["Account Manager","Contact","Product"],values=["Sale Price","Licenses"],
               aggfunc='sum',fill_value=0,margins=True, margins_name='Total', dropna=True)
```
