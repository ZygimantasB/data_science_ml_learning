```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pathlib import Path

from IPython.core.pylabtools import figsize
from pyspark.pandas.mlflow import load_model
```

```python
df_path = Path.cwd() / '..' / '..' / 'data' / 'Advertising.csv'
```

```python
df = pd.read_csv(df_path)
```

```python
df.head()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6))

axes[0].plot(df['TV'], df['sales'], 'o')
axes[0].set_ylabel('Sales')
axes[0].set_title('TV Spend')

axes[1].plot(df['radio'], df['sales'], 'o')
axes[1].set_ylabel('Sales')
axes[1].set_title('Radio')

axes[2].plot(df['newspaper'], df['sales'], 'o')
axes[2].set_ylabel('Sales')
axes[2].set_title('Newspaper')
```

```python
sns.pairplot(data=df)
```

```python
df.head()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
test_predictions = model.predict(X_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
test_predictions
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
MSE = mean_squared_error(y_test, test_predictions)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
```

```python
RMSE = np.sqrt(y_test, test_predictions)
```

```python
print(f'MAE {MAE}, MSE {MSE}, RMSE {RMSE}')
```
