```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
```

```python
symbolic_link = Path("../data/Advertising.csv")
directory = symbolic_link.resolve()

df = pd.read_csv(directory)
```

```python
df
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
# sns.scatterplot(x='total_spend', y='sales', hue='total_spend', data=df)
sns.scatterplot(x='total_spend', y='sales', data=df)
```

```python
sns.regplot(x='total_spend', y='sales', data=df, ci=None)
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='total_spend', y='sales', data=df)
plt.title("Scatter Plot of Total Spend vs Sales")
plt.xlabel("Total Spend on Advertising")
plt.ylabel("Sales")
plt.show()

# Regression plot
plt.figure(figsize=(10, 5))
sns.regplot(x='total_spend', y='sales', data=df, ci=None)
plt.title("Regression Plot of Total Spend vs Sales")
plt.xlabel("Total Spend on Advertising")
plt.ylabel("Sales")
plt.show()
```
