```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_path = r'F:\DataSpell\data_science_ml_learning\DATA\Advertising.csv'
```

```python
df = pd.read_csv(df_path)
```

```python
df.head()
```

```python
X = df.drop('sales', axis=1)
```

```python
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter = PolynomialFeatures(degree=3, include_bias=False)
```

```python
poly_features = polynomial_converter.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, train_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MSE = mean_squared_error(y_test, test_predictions)
RMSE = np.sqrt(MSE)
```

```python
print(f'MAE: {MAE} \nMSE: {MSE} \nRMSE: {RMSE}')
```

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0), scoring='neg_mean_absolute_error')
```

```python
ridge_cv_model.fit(X_train, y_train)
```

```python
ridge_cv_model.alpha_
```

```python
test_predictions = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MSE = mean_squared_error(y_test, test_predictions)
RMSE = np.sqrt(MSE)
```

```python
print(f'MAE: {MAE} \nMSE: {MSE} \nRMSE: {RMSE}')
```

```python
pd.DataFrame(ridge_cv_model.coef_, columns=['Ridge Coefficients'])
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MSE = mean_squared_error(y_test, test_predictions)
RMSE = np.sqrt(MSE)
```
