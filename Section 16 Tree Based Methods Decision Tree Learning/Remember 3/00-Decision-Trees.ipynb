```python
X = pd.get_dummies(df.drop('species',axis=1),drop_first=True)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
base_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
pd.DataFrame(confusion_matrix(base_pred, y_test), index=model.classes_, columns=model.classes_)
```

```python
ConfusionMatrixDisplay(confusion_matrix(base_pred, y_test), display_labels=model.classes_).plot(cmap='Blues');
```

```python
print(classification_report(y_test,base_pred))
```

```python
model.feature_importances_
```

```python
pd.Series(model.feature_importances_, index=X.columns).sort_values().plot(kind='bar');
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance'])
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12, 8))
plot_tree(model);
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(model,filled=True,feature_names=X.columns);
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test, model_preds))
    print('\n')
    plt.figure(figsize=(12, 8), dpi=150)
    plot_tree(model, filled=True, feature_names=X.columns)
```

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
pruned_tree = DecisionTreeClassifier(max_leaf_nodes=3)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
