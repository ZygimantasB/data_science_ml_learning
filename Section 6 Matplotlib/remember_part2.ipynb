# 02-Matplotlib-SubPlots

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, color='r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for axes in axes:
  axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a, b)
axes[1].plot(x, -y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[1][1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x, y)
axes[0][1].plot(a, b)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, dpi=250, figsize=(8, 6))

axes[0][0].plot(x, y)
axes[0][0].set_title('0 0 Title')

axes[0][1].plot(a, b)
axes[0][1].set_title('0 1 Title')

axes[1][0].plot(b, a)
axes[1][0].set_title('1 0 Title')

axes[1][1].plot(y, x)
axes[1][1].set_title('1 1 Title')

fig.suptitle('Figure Level Title', fontsize=16)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, dpi=250, figsize=(8, 6))

axes[0][0].plot(a, b)
axes[0][1].plot(x, y)
axes[1][0].plot(b, a)
axes[1][1].plot(y, x)

fig.subplots_adjust(left=None,
                    bottom=None,
                    right=None,
                    top=None,
                    wspace=0.5,
                    hspace=0.1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, dpi=250, figsize=(8, 6))

axes[0][0].plot(a, b)
axes[0][1].plot(x, y)
axes[1][0].plot(b, a)
axes[1][1].plot(y, x)

fig.savefig('my_picture.png', dpi=200, bbox_inches='tight')
```

# 03-Matplotlib-Styling-Plots

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**3, label='x**3')
ax.plot(x, x**4, label='x**4')
ax.plot(x, x**2, label='x**2')
ax.plot(x, x**1, label='x**1')

ax.legend(loc=1)
ax.legend(loc=2)
ax.legend(loc=3)
ax.legend(loc=4)
ax.legend(loc=0)

ax.legend(loc=(1.1, 0.5))
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3, 'g--')
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color='blue', alpha=0.5)
ax.plot(x, x+2, color='#8B008B')
ax.plot(x, x+3, color='#FF8C00')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, x+1, color='red', linewidth=0.25)
ax.plot(x, x+2, color='red', linewidth=0.50)
ax.plot(x, x+3, color='r', lw='1')
ax.plot(x, x+4, color='r', lw='2')
```

```python
fig, ax = plt.subplots()

ax.plot(x, x-1, color='green', lw=3, linestyle='-')
ax.plot(x, x-2, color='g', lw=3, ls='-.')
ax.plot(x, x-3, color='blue', lw=3, ls=':')
ax.plot(x, x-4, color='b', lw=3, ls='--')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

lines = ax.plot(x, x)

print(type(lines))
```

```python
fig, ax = plt.subplots()

lines = ax.plot(x, x+8, color='black', lw=1.5)
lines[0].set_dashes([10, 10])
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

lines = ax.plot(x, x+8, color='black', lw=1.5)
lines[0].set_dashes([1, 1, 1, 1, 10, 10])
```
